```python
class JointDetectionNet(nn.Module):
    def __init__(self, channel_in):
        super(JointDetectionNet, self).__init__()
        self.init = nn.Sequential(
            nn.Conv2d(channel_in, 64, 
                      kernel_size=7, stride=3, 
                      padding=3),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            ResBlock(64, 128),
            nn.MaxPool2d(2,2),
            ResBlock(128, 128),
            ResBlock(128, 128),
            ResBlock(128, 256)
        )
        self.hg1 = Hourglass(4, 256, 512)
        
    def forward(self, x):
        return self.init(x)
```
